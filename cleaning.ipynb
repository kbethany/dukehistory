```python
import pandas as pd

# Path to your Excel file
file_path = 'Duke History Grads IN PROGRESS.xlsx'

# Define the sheet names you want to include
year_sheets = [str(year) for year in range(2016, 2024)]  # 2024 not included

# Read and append only the specified sheets
all_years_df = pd.concat(
    [pd.read_excel(file_path, sheet_name=sheet) for sheet in year_sheets if sheet in pd.ExcelFile(file_path).sheet_names],
    ignore_index=True
)

# Save the combined DataFrame to CSV
all_years_df.to_csv('2016to2023grads.csv', index=False)

all_years_df.head()
```

```python
# drop fields in the dataframe
columns_to_drop = ['Student ID', 'Duke Unique ID', 'Primary Name', 'Degree/Diploma Name']

# Drop them from the combined DataFrame (ignore errors if columns are missing)
all_years_df = all_years_df.drop(columns=columns_to_drop, errors='ignore')

all_years_df.head()
```

```python
# stripped data to new csv
all_years_df.to_csv('2016to2023grads-noids.csv', index=False)
```
